```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
```

```python
data = pd.read_csv("housing.csv")

print("\n--- Before Dropping Nulls ---")
data.info()

```

```python
data.dropna(inplace=True)
print("\n--- After Dropping Nulls ---")
data.info()
```

```python
x = data.drop(['median_house_value'], axis=1)
y = data['median_house_value']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
train_data = x_train.join(y_train)
train_data.head()
```

```python
for col in ['total_rooms', 'total_bedrooms', 'population', 'households']:
    train_data[col] = np.log(train_data[col] + 1)
train_data.head()
```

```python
dummies = pd.get_dummies(train_data['ocean_proximity'], prefix='ocean')
train_data = pd.concat([train_data.drop('ocean_proximity', axis=1), dummies], axis=1)
train_data.head()
```

```python
train_data['bedroom_ratio'] = train_data['total_bedrooms'] / train_data['total_rooms']
train_data['household_rooms'] = train_data['total_rooms'] / train_data['households']
train_data.head()
```

```python
plt.figure(figsize=(15, 8))
sns.heatmap(train_data.corr(numeric_only=True), annot=True, cmap='YlGnBu')
plt.title("Correlation Heatmap - Training Data")
plt.show()
```

```python
x_train = train_data.drop(['median_house_value'], axis=1)
y_train = train_data['median_house_value']

reg = LinearRegression()
reg.fit(x_train, y_train)
```

```python
# Join x_test and y_test
test_data = x_test.join(y_test)

# Apply log transformation safely
for col in ['total_rooms', 'total_bedrooms', 'population', 'households']:
    if col in test_data.columns:
        test_data[col] = np.log(test_data[col] + 1)

# One-hot encode ocean_proximity only if it exists
if 'ocean_proximity' in test_data.columns:
    dummies = pd.get_dummies(test_data['ocean_proximity'], prefix='ocean')
    test_data = pd.concat([test_data.drop('ocean_proximity', axis=1), dummies], axis=1)

# Add engineered features safely
if 'total_bedrooms' in test_data.columns and 'total_rooms' in test_data.columns:
    test_data['bedroom_ratio'] = test_data['total_bedrooms'] / test_data['total_rooms']

if 'total_rooms' in test_data.columns and 'households' in test_data.columns:
    test_data['household_rooms'] = test_data['total_rooms'] / test_data['households']

# Align columns with train data
test_data = test_data.reindex(columns=train_data.columns, fill_value=0)

# Separate x_test and y_test
x_test = test_data.drop(['median_house_value'], axis=1)
y_test = test_data['median_house_value']
```

```python
print("Linear Regression Test Score:", reg.score(x_test, y_test))
```

```python
forest = RandomForestRegressor(random_state=42)
forest.fit(x_train, y_train)

print("Random Forest Train Score:", forest.score(x_train, y_train))
print("Random Forest Test Score:", forest.score(x_test, y_test))
```

```python
param_grid = {
    "n_estimators": [3, 10, 30],
    "max_features": [2, 4, 6, 8]
}

grid_search = GridSearchCV(
    RandomForestRegressor(random_state=42),
    param_grid,
    cv=5,
    scoring="neg_mean_squared_error",
    return_train_score=True,
    n_jobs=-1
)

grid_search.fit(x_train, y_train)
```

```python
best_forest = grid_search.best_estimator_

print("\nBest Parameters from Grid Search:", grid_search.best_params_)
print("Best Random Forest Test Score:", best_forest.score(x_test, y_test))
```
